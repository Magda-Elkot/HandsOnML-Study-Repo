# What Is Machine Learning?  

Machine Learning is the science of programming computers to learn from data.  

## Definitions  
- **Arthur Samuel (1959):**  
  *Machine Learning is the field of study that gives computers the ability to learn without being explicitly programmed.*  

- **Tom Mitchell (1997):**  
  *A computer program learns from experience (E) with respect to some task (T) and performance measure (P) if its performance on T improves with experience E.*  

## Example: Spam Filter  
- **Task (T):** Flagging spam emails.  
- **Experience (E):** Training data (labeled spam and non-spam emails).  
- **Performance Measure (P):** Accuracy (ratio of correctly classified emails).  

## Key Concept  
Downloading large amounts of data (e.g., Wikipedia) does not mean a system has learned; learning involves improving performance on a specific task.  


# Why Use Machine Learning?  

### Traditional Approach to Spam Filtering  
1. Identify common spam characteristics (e.g., words like “4U,” “credit card,” “free,” “amazing”).  
2. Write detection rules based on observed patterns in the subject, sender, and email body.  
3. Test the program and refine rules iteratively until performance is satisfactory.

### Limitations of the Traditional Approach  
- The program becomes a long list of complex rules, making it difficult to maintain.  

### Advantages of Machine Learning  
- Automatically learns patterns that indicate spam.  
- Detects frequent words/phrases in spam compared to ham.  
- Results in a shorter, easier-to-maintain, and more accurate program.

#### Limitations of Traditional Programming  
- Spammers can bypass fixed rules by slightly modifying words (e.g., changing “4U” to “For U”).  
- Requires constant manual updates to keep up with new spam techniques.  

#### Advantages of Machine Learning  
- Automatically detects new spam patterns based on user feedback.  
- Adapts to evolving spam tactics without manual intervention.

### Machine Learning for Complex Problems

- Some problems are too complex or lack a known algorithm (e.g., speech recognition).  
- Hardcoded rules do not scale to thousands of words, different speakers, and noisy environments.  
- Machine Learning enables models to learn from examples instead of relying on predefined rules.  

#### Machine Learning for Human Learning  
- ML models can be analyzed to understand what they have learned.  
- Example: A trained spam filter can reveal key words and patterns associated with spam.

### **Reinforcement Learning**

### **Batch and Online Learning**  

#### **Batch Learning**  
- Learns from all available data at once (**offline learning**).  
- Requires **retraining from scratch** when new data is available.  
- Suitable for **static datasets**, but costly for rapidly changing data.  

#### **Online Learning**  
- Learns **incrementally** from a stream of data.  
- Efficient for **real-time adaptation** (e.g., stock price prediction).  
- Suitable for **large datasets** that cannot fit in memory (**out-of-core learning**).

### **Instance-Based vs. Model-Based Learning**  

#### **Instance-Based Learning**  
- Memorizes training examples and classifies new data based on **similarity** to known instances.  
- Example: A spam filter that flags emails **similar** to known spam.

#### **Model-Based Learning**  
- Builds a **predictive model** from training data and **generalizes** to new instances.  
- Example: Predicting **happiness based on income** using a regression model.

### **Model Selection and Training**  

1. **Selecting a Model**  
   - Choose a model that best represents the data (e.g., **linear regression** for GDP vs. life satisfaction).  

2. **Training the Model**  
   - Adjust parameters (**θ0, θ1**) to minimize error using a **cost function**.  
   - Example: **Linear Regression** finds the best-fit line for predicting life satisfaction from GDP.  

3. **Making Predictions**  
   - Once trained, the model can predict outcomes for new data (e.g., **Cyprus’s life satisfaction**).  

4. **Instance-Based Alternative**  
   - **k-Nearest Neighbors (k-NN) regression** predicts values based on the closest known data points.  

5. **Improving Predictions**  
   - Use more **attributes** (e.g., employment, health).  
   - Get **better quality data**.  
   - Choose a **more complex model** if needed.  

6. **Finally, you applied the model to make predictions on new cases (this is called inference), hoping
that this model will generalize well.**

This workflow—**studying data, selecting a model, training, and making predictions**—is at the core of **Machine Learning projects**.  


### **Main Challenges of Machine Learning**  

1. **Bad Algorithm or Bad Data**  
   - Errors stem from **poor algorithm choice** or **low-quality data**.  

2. **Insufficient Training Data**  
   - ML models require **large datasets** to generalize well.  
   - Complex tasks like **image or speech recognition** need **millions of examples** unless leveraging pre-trained models.  

3. **Nonrepresentative Training Data**  
- A training set must **accurately represent** the real-world cases it will be applied to.  
- A flawed dataset can cause **sampling noise** (random nonrepresentative data) or **sampling bias** (systematic misrepresentation).  

4. **Poor-Quality Data**  
- Errors, missing values, outliers, and noise in training data **reduce model accuracy**.  
- Data cleaning is essential:  
  - Remove or correct **outliers**.  
  - Handle **missing values** (e.g., use median replacement).  

5. **Irrelevant Features**  
- Including **unnecessary** features can degrade model performance.  
- **Feature engineering** improves learning:  
  - **Feature selection**: Choosing only the most relevant features.  
  - **Feature extraction**: Creating new, more informative features.  
  - **Gathering additional data** for better feature representation.  

6. **Overfitting the Training Data**  
- **Overfitting** happens when a model learns the **noise** instead of real patterns, making it perform poorly on new data.  
- Example: A model detects a meaningless correlation (e.g., countries with a "w" in their name have higher life satisfaction).  
- **Solutions to Overfitting**:  
  - Use **simpler models** (e.g., linear regression instead of high-degree polynomials).  
  - Gather **more training data** to reduce noise.  
  - Apply **regularization** to prevent the model from adapting too closely to training data.  

7. **Underfitting the Training Data**  
- **Underfitting** happens when a model is **too simple** to capture the underlying structure of data.  
- Example: A linear model predicting life satisfaction may be too simplistic.  
- **Solutions to Underfitting**:  
  - Use a **more complex model** with additional parameters.  
  - Perform **feature engineering** to enhance data quality.  
  - Reduce **regularization constraints** to allow better fitting.  

---

8. **Stepping Back: The Big Picture**  
Now that we’ve covered the challenges, let’s step back and look at Machine Learning as a whole:  

- **Machine Learning automates learning from data** instead of relying on explicitly coded rules.  
- There are different **types of ML systems**:  
  - **Supervised vs. Unsupervised**  
  - **Batch vs. Online Learning**  
  - **Instance-Based vs. Model-Based**  
- A Machine Learning project involves:  
  - **Gathering and preparing data** for training.  
  - **Selecting and training a model** to fit the data.  
  - **Evaluating and fine-tuning the model** to improve its generalization to new cases.  
- The system will fail if:  
  - The **training data is insufficient, unrepresentative, noisy, or irrelevant**.  
  - The model is **too simple (underfitting) or too complex (overfitting)**.

## **Final Thought**  
Once a model is trained, we cannot just “hope” it generalizes well. It must be **evaluated and fine-tuned** to ensure reliable performance in real-world applications.  


# **Testing and Validating Machine Learning Models**

## **Why Testing and Validation Matter**
The only way to ensure a model generalizes well to new data is to test it on unseen cases. One approach is to deploy the model in a **real-world scenario** and monitor its performance. However, if the model performs poorly, users may face issues—so this is not always ideal.

Instead, a more controlled approach involves **splitting data into training and test sets**.

---

## **1. Training and Test Sets**
- The dataset is divided into:
  - **Training Set** (typically **80%**): Used to train the model.
  - **Test Set** (typically **20%**): Used to estimate the **generalization error** (error on unseen data).
- If a model has **low training error** but **high test error**, it is **overfitting**.

---

## **2. Model Selection and Hyperparameter Tuning**
- When comparing multiple models (e.g., **linear vs. polynomial regression**), we evaluate their performance using the test set.
- Hyperparameter tuning (e.g., adjusting **regularization**) requires evaluating multiple models.
- **Problem**: If we repeatedly adjust the model based on test set performance, the test set effectively becomes part of the training process, leading to **biased results**.

---

## **3. Validation Set and Cross-Validation**
### **a. The Validation Set Approach**
- Instead of relying solely on a test set, we introduce a **validation set**:
  - **Training Set**: Used to train models.
  - **Validation Set**: Used to tune hyperparameters and select the best model.
  - **Test Set**: Used **only once** for final evaluation.
  
### **b. Cross-Validation**
- Instead of reserving a fixed validation set (which reduces the amount of training data), we use **cross-validation**:
  - The training set is **split into multiple subsets**.
  - The model is trained on different combinations and validated on the remaining parts.
  - This ensures the model is evaluated across different data partitions.
- After tuning, a **final model is trained** on the full training set and tested once on the test set.

---

## **Final Takeaway**
Testing and validation ensure that a model **performs well on new data** and prevents overfitting. Using techniques like a **validation set and cross-validation**, we can **select the best model** and **fine-tune hyperparameters** without compromising the test set’s integrity.

### 1. How would you define Machine Learning?  
Machine Learning is a field of artificial intelligence where algorithms learn from data, allowing them to make predictions or decisions without being explicitly programmed. It relies on finding patterns in data and uses these patterns to make inferences about new data.

### 2. Can you name four types of problems where it shines?  
Machine Learning excels in:  
- **Image and speech recognition** (e.g., identifying faces in photos)  
- **Natural language processing** (e.g., translating languages, generating text)  
- **Predictive analytics** (e.g., forecasting sales, predicting customer churn)  
- **Anomaly detection** (e.g., fraud detection, defect detection in manufacturing)  

### 3. What is a labeled training set?  
A labeled training set is a dataset where each data point is paired with an output label. In supervised learning, this dataset is used to teach the model.

### 4. What are the two most common supervised tasks?  
The two most common supervised tasks are:  
- **Classification** (assigning labels to categories)  
- **Regression** (predicting continuous values)  

### 5. Can you name four common unsupervised tasks?  
Common unsupervised tasks include:  
- **Clustering** (grouping data points into clusters)  
- **Anomaly detection** (identifying unusual data points)  
- **Dimensionality reduction** (simplifying data by reducing its number of features)  
- **Association** (finding patterns or associations in data, such as in market basket analysis)  

### 6. What type of Machine Learning algorithm would you use to allow a robot to walk in various unknown terrains?  
A **reinforcement learning** algorithm would be appropriate, as it allows the robot to learn through trial and error by interacting with the environment and receiving feedback in the form of rewards or penalties.

### 7. What type of algorithm would you use to segment your customers into multiple groups?  
A **clustering algorithm**, which is unsupervised, would be used to group customers based on similarities without predefined labels.

### 8. Would you frame the problem of spam detection as a supervised or an unsupervised learning problem?  
**Supervised learning** is typically used for spam detection because we have labeled examples of both spam and non-spam messages to train the model.

### 9. What is an online learning system?  
An **online learning system** is one that continuously learns as new data comes in, rather than learning in one big batch. It’s useful for applications where data is constantly updated, such as stock price prediction or recommendation systems.

### 10. What is out-of-core learning?  
**Out-of-core learning** refers to training models on data that is too large to fit into memory by processing it in small chunks. This approach is useful for handling big data on limited hardware.

### 11. What type of learning algorithm relies on a similarity measure to make predictions?  
**Instance-based learning algorithms**, like **k-Nearest Neighbors (k-NN)**, rely on similarity measures to make predictions by comparing new instances to known instances.

### 12. What is the difference between a model parameter and a learning algorithm’s hyperparameter?  
- **Model parameters** are learned from the data during training and define how the model makes predictions (e.g., weights in a neural network).  
- **Hyperparameters** are settings specified before training that control the learning process (e.g., learning rate, regularization strength) but aren’t adjusted through training.

### 13. What do model-based learning algorithms search for? What is the most common strategy they use to succeed? How do they make predictions?  
Model-based learning algorithms search for an **optimal model** that generalizes well to unseen data by minimizing a **cost function** (e.g., Mean Squared Error in regression). They usually use an **optimization technique**, like **gradient descent**, to find the optimal parameters. Once trained, they use the learned model parameters to make predictions on new data.

### 14. Can you name four of the main challenges in Machine Learning?  
- **Insufficient or poor-quality data**  
- **Overfitting** (model performs well on training data but poorly on new data)  
- **Underfitting** (model is too simple and performs poorly even on training data)  
- **Data bias** (the training data isn’t representative of the real-world scenario)  

### 15. If your model performs great on the training data but generalizes poorly to new instances, what is happening? Can you name three possible solutions?  
This is called **overfitting**. Possible solutions include:  
- Simplifying the model by reducing its complexity  
- Collecting more training data  
- Using **regularization** techniques like **L1** or **L2** regularization to penalize overly complex models  

### 16. What is a test set and why would you want to use it?  
A **test set** is a dataset set aside to evaluate the model’s performance on new, unseen data. It helps assess the model’s ability to generalize beyond the training data.

### 17. What is the purpose of a validation set?  
A **validation set** is used during model training to **tune hyperparameters** and **evaluate model performance** before final testing. It helps in selecting the best model without overfitting to the training data.

### 18. What can go wrong if you tune hyperparameters using the test set?  
Using the test set for tuning can lead to **overfitting on the test data**, making the model perform well on that specific dataset but poorly on truly new data.

### 19. What is cross-validation and why would you prefer it to a validation set?  
**Cross-validation** is a technique where the data is split into multiple subsets (**folds**), and the model is trained and validated on different folds. This provides a more **robust estimate** of model performance by averaging results across folds and allows **better use of the data** compared to a simple validation split.

# Load and prepare Life satisfaction data
To create lifesat.csv, I downloaded the Better Life Index (BLI) data from OECD's website (to get the Life Satisfaction for each country), and World Bank GDP per capita data from OurWorldInData.org. The BLI data is in datasets/lifesat/oecd_bli.csv (data from 2020), and the GDP per capita data is in datasets/lifesat/gdp_per_capita.csv (data up to 2020).

If you want to grab the latest versions, please feel free to do so. However, there may be some changes (e.g., in the column names, or different countries missing data), so be prepared to have to tweak the code.

```python

import urllib.request

datapath = Path() / "datasets" / "lifesat"
datapath.mkdir(parents=True, exist_ok=True)

data_root = "https://github.com/ageron/data/raw/main/"
for filename in ("oecd_bli.csv", "gdp_per_capita.csv"):
    if not (datapath / filename).is_file():
        print("Downloading", filename)
        url = data_root + "lifesat/" + filename
        urllib.request.urlretrieve(url, datapath / filename)
```

```python
oecd_bli = pd.read_csv(datapath / "oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

Preprocess the GDP per capita data to keep only the year 2020:

```python
gdp_year = 2020
gdppc_col = "GDP per capita (USD)"
lifesat_col = "Life satisfaction"

gdp_per_capita = gdp_per_capita[gdp_per_capita["Year"] == gdp_year]
gdp_per_capita = gdp_per_capita.drop(["Code", "Year"], axis=1)
gdp_per_capita.columns = ["Country", gdppc_col]
gdp_per_capita.set_index("Country", inplace=True)

gdp_per_capita.head()
```

Preprocess the OECD BLI data to keep only the Life satisfaction column:

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")

oecd_bli.head()
```

Now let's merge the life satisfaction data and the GDP per capita data, keeping only the GDP per capita and Life satisfaction columns:

```python
full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                              left_index=True, right_index=True)
full_country_stats.sort_values(by=gdppc_col, inplace=True)
full_country_stats = full_country_stats[[gdppc_col, lifesat_col]]

full_country_stats.head()
```


To illustrate the risk of overfitting, I use only part of the data in most figures (all countries with a GDP per capita between min_gdp and max_gdp). Later in the chapter I reveal the missing countries, and show that they don't follow the same linear trend at all.

```python
min_gdp = 23_500
max_gdp = 62_500

country_stats = full_country_stats[(full_country_stats[gdppc_col] >= min_gdp) &
                                   (full_country_stats[gdppc_col] <= max_gdp)]
country_stats.head()
```

```python
country_stats.to_csv(datapath / "lifesat.csv")
full_country_stats.to_csv(datapath / "lifesat_full.csv")
```

```python
from sklearn import preprocessing
from sklearn import pipeline

full_country_stats.plot(kind='scatter', figsize=(8, 3),
                        x=gdppc_col, y=lifesat_col, grid=True)

poly = preprocessing.PolynomialFeatures(degree=10, include_bias=False)
scaler = preprocessing.StandardScaler()
lin_reg2 = linear_model.LinearRegression()

pipeline_reg = pipeline.Pipeline([
    ('poly', poly),
    ('scal', scaler),
    ('lin', lin_reg2)])
pipeline_reg.fit(Xfull, yfull)
curve = pipeline_reg.predict(X[:, np.newaxis])
plt.plot(X, curve)

plt.axis([0, 115_000, min_life_sat, max_life_sat])

save_fig('overfitting_model_plot')
plt.show()
```
